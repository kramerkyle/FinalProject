```python
import pandas as pd
```

```python
# Load in the data set.
file = 'Life_Expectancy.csv'
life_df = pd.read_csv(file)
life_df
```

```python
# Rename columns to match SQL database
life_expectancy_df = life_df.rename(columns = {'State':'state', 'County': 'county', 'Census Tract Number': 'census_tract_number', 'Life Expectancy':'life_expectancy', 'Life Expectancy Range': 'life_expectancy_range', 'Life Expectancy Standard Error': 'life_expectancy_standard_error'})
life_expectancy_df
```

```python
# Set state as index
life_expectancy_df.set_index('state', inplace=True)
life_expectancy_df
```

```python
#Drop null values
life_expectancy_df.dropna(axis=0, how='any', inplace=True)
life_expectancy_df
```

```python
# Export to a CSV
life_expectancy_df.to_csv('life_expectancy_clean_2.csv', header=False)
```
